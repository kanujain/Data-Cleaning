```python
Loc[]:
    >> works on column range i.e. POSITION
    >> i.e 0, 1, 2...default values in the data frame as indexes

iloc[]:
    >> When any column attribute is set as indexes
        >> data.set_index('id', inplace = True)  => Stting 'id' as index 
    >> Here iloc works 
    >> loc would not be helpful to navigate in this situation till the time we are not familier with id values

data.iloc[0:3, :]
        >> we'll get the results of first 3 rows 

data.loc[0:3, :]
        >> we get no result as the index is set by ID not range

IMP:
For normal dataframe wrt position
    data.loc[0:3, :] => returns row from 0 to 3  #range is inclusive 
    data.iloc[0:3, :] => returns row from 0 to 2 #range is exclusive
```
